```python
#will convert standard chess format of position from board to an array indices format i.e [row, col]
def convertPos(pos):
  toreturn = [-1,-1]
  #second number indicates row
  toreturn[0] = int(pos[1])-1
  #first letter indicates column
  for each in colnames:
    if colnames[each] == pos[0]:
      toreturn[1] = each
      break
  return toreturn

#will return all possible moves according to the depth given
def GetLegalMoves(pos, board, depth):
  pc = board.b[pos[0]][pos[1]]
  legalmoves = list()
  oppcolor = "-"
  if pc[0][0] == 'W':
    oppcolor = 'B'
    owncolor = 'W'
  else:
    oppcolor = 'W'
    owncolor = 'B'

  #QUEEN MOVES----------------------------------------------------------------------
  if pc[0][1] == 'Q':
    i = pos[0]  #row 
    j = pos[1]  #col
    tempdepth = depth
    while j < 7 and tempdepth > 0:                #move right
      tempdepth -= 1
      j += 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j]) 
    j = pos[1]  #reset col
    tempdepth = depth
    while j > 0 and tempdepth > 0:             #move left
      j -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j]) 
    j = pos[1]
    tempdepth = depth
    while i > 0 and tempdepth > 0:             #move up
      i -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    tempdepth = depth
    while i < 7 and tempdepth > 0:              #move down
      i += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    tempdepth = depth
    while i > 0 and j > 0 and tempdepth > 0:    #move NW
      i -= 1
      j -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i < 7 and j > 0 and tempdepth > 0:    #move SW
      i += 1
      j -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i > 0 and j < 7 and tempdepth > 0:    #move NE
      i -= 1
      j += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i < 7 and j < 7 and tempdepth > 0:    #move SE
      i += 1
      j += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:
        break
      elif board.b[i][j][0][0] == oppcolor:
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])

  #BISHOP MOVES----------------------------------------------------------------------
  elif pc[0][1] == 'B':
    i = pos[0]  #row 
    j = pos[1]  #col
    tempdepth = depth
    while i > 0 and j > 0 and tempdepth > 0:    #move NW
      tempdepth -= 1
      i -= 1
      j -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i < 7 and j > 0 and tempdepth > 0:    #move SW
      i += 1
      j -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i > 0 and j < 7 and tempdepth > 0:    #move NE
      i -= 1
      j += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    j = pos[1] 
    tempdepth = depth
    while i < 7 and j < 7 and tempdepth > 0:    #move SE
      i += 1
      j += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])

  #ROOK MOVES----------------------------------------------------------------------
  elif pc[0][1] == 'R':
    i = pos[0]  #row 
    j = pos[1]  #col
    tempdepth = depth
    while j < 7 and tempdepth > 0:                #move right
      j += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j]) 
    j = pos[1]  #reset col
    tempdepth = depth
    while j > 0 and tempdepth > 0:             #move left
      j -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    j = pos[1]
    tempdepth = depth
    while i > 0 and tempdepth > 0:             #move up
      i -= 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])
    i = pos[0]
    tempdepth = depth
    while i < 7 and tempdepth > 0:              #move down
      i += 1
      tempdepth -= 1
      if board.b[i][j][0][0] == owncolor:         #cannot cross your own color
        break
      elif board.b[i][j][0][0] == oppcolor:       #can take the first opponent piece in your path but then stop moving further
        legalmoves.append([i,j])
        break
      else:
        legalmoves.append([i,j])

  #KNIGHT MOVES----------------------------------------------------------------------
  elif pc[0][1] == 'N':
    i = pos[0]  #row 
    j = pos[1]  #col
    if (i + 2) < 8 and (j + 1) < 8 and board.b[i+2][j+1][0][0] != owncolor:
        legalmoves.append([i+2,j+1])
    if (i + 2) < 8 and (j - 1) > -1 and board.b[i+2][j-1][0][0] != owncolor:
        legalmoves.append([i+2,j-1])
    if (i - 2) > -1 and (j + 1) < 8 and board.b[i-2][j+1][0][0] != owncolor:
        legalmoves.append([i-2,j+1])
    if (i - 2) > -1 and (j - 1) > -1 and board.b[i-2][j-1][0][0] != owncolor:
        legalmoves.append([i-2,j-1])
    if (j + 2) < 8 and (i + 1) < 8 and board.b[i+1][j+2][0][0] != owncolor:
        legalmoves.append([i+1,j+2])
    if (j + 2) < 8 and (i - 1) > -1 and board.b[i-1][j+2][0][0] != owncolor:
        legalmoves.append([i-1,j+2])
    if (j - 2) > -1 and (i + 1) < 8 and board.b[i+1][j-2][0][0] != owncolor:
        legalmoves.append([i+1,j-2])
    if (j - 2) > -1 and (i - 1) > -1 and board.b[i-1][j-2][0][0] != owncolor:
        legalmoves.append([i-1,j-2])

  #PAWN MOVES----------------------------------------------------------------------
  #Black pawn - can only move down
  elif pc[0] == 'BP':
    i = pos[0]  #row 
    j = pos[1]  #col
    if i == 1:
      if i+1 < 8 and j-1 > -1 and board.b[i+1][j-1][0][0] == 'W':
        legalmoves.append([i+1,j-1])
      if i+1 < 8 and j+1 < 8 and board.b[i+1][j+1][0][0] == 'W':
        legalmoves.append([i+1,j+1])
      if board.b[i+1][j] == EM:
        i += 1
        legalmoves.append([i,j])
        if board.b[i+1][j] == EM:
          i += 1
          legalmoves.append([i,j])
    else:
      if i+1 < 8 and board.b[i+1][j] == EM:
        legalmoves.append([i+1,j])
      if i+1 < 8 and j-1 > -1 and board.b[i+1][j-1][0][0] == 'W':
        legalmoves.append([i+1,j-1])
      if i+1 < 8 and j+1 < 8 and board.b[i+1][j+1][0][0] == 'W':
        legalmoves.append([i+1,j+1])
  #white pawn - can only move up
  elif pc[0] == 'WP':
    i = pos[0]  #row 
    j = pos[1]  #col
    if i == 6:
      if i-1 > -1 and j-1 > -1 and board.b[i-1][j-1][0][0] == 'B':
        legalmoves.append([i-1,j-1])
      if i-1 > -1 and j+1 < 8 and board.b[i-1][j+1][0][0] == 'B':
        legalmoves.append([i-1,j+1])
      if board.b[i-1][j] == EM:
        i -= 1
        legalmoves.append([i,j])
        if board.b[i-1][j] == EM:
          i -= 1
          legalmoves.append([i,j])
    else:
      if i-1 > -1 and board.b[i-1][j] == EM:
        legalmoves.append([i-1,j])
      if i-1 > -1 and j-1 > -1 and board.b[i-1][j-1][0][0] == 'B':
        legalmoves.append([i-1,j-1])
      if i-1 > -1 and j+1 < 8 and board.b[i-1][j+1][0][0] == 'B':
        legalmoves.append([i-1,j+1])

  #KING MOVES----------------------------------------------------------------------
  elif pc[0][1] == 'K':
    i = pos[0]  #row 
    j = pos[1]  #col
    if i + 1 < 8 and board.b[i+1][j][0][0] != owncolor and CanBeCaptured(pc,[i+1,j],board)==False:
      legalmoves.append([i+1,j])
    if i - 1 > -1 and board.b[i-1][j][0][0] != owncolor and CanBeCaptured(pc,[i-1,j],board)==False:
      legalmoves.append([i-1,j])
    if j + 1 < 8 and board.b[i][j+1][0][0] != owncolor and CanBeCaptured(pc,[i,j+1],board)==False:
      legalmoves.append([i,j+1])
    if j - 1 > -1 and board.b[i][j-1][0][0] != owncolor and CanBeCaptured(pc,[i,j-1],board)==False:
      legalmoves.append([i,j-1])
    if i + 1 < 8 and j + 1 < 8 and board.b[i+1][j+1][0][0] != owncolor and CanBeCaptured(pc,[i+1,j+1],board)==False:
      legalmoves.append([i+1,j+1])
    if i + 1 < 8 and j - 1 > -1 and board.b[i+1][j-1][0][0] != owncolor and CanBeCaptured(pc,[i+1,j-1],board)==False:
      legalmoves.append([i+1,j-1])
    if i - 1 > -1 and j + 1 < 8 and board.b[i-1][j+1][0][0] != owncolor and CanBeCaptured(pc,[i-1,j+1],board)==False:
      legalmoves.append([i-1,j+1])
    if i - 1 > -1 and j - 1 > -1 and board.b[i-1][j-1][0][0] != owncolor and CanBeCaptured(pc,[i-1,j-1],board)==False:
      legalmoves.append([i-1,j-1])
    if pos[0] == 7 and pos[1] == 4:
      if board.castlingboolwhite == True and owncolor == 'W' and CanBeCaptured(pc,[i,j-2],board)==False:
        if board.b[i][j-1] == EM and board.b[i][j-2] == EM and board.b[i][j-3] == EM and board.b[i][j-4] == WR:
          legalmoves.append([i,j-2])
      if board.castlingboolwhite == True and owncolor == 'W' and CanBeCaptured(pc,[i,j+2],board)==False:
        if board.b[i][j+1] == EM and board.b[i][j+2] == EM and board.b[i][j+3] == WR:
          legalmoves.append([i,j+2])
      

  return legalmoves

```

```python
#will simply return true or false if moving to a new position would capture opponents piece
def CanCaptureOpponent(prevpos, newpos, board):
  owncolor = board.b[prevpos[0]][prevpos[1]][0][0]
  if owncolor == 'W':
    oppcolor = 'B'
  elif owncolor == 'B':
    oppcolor = 'W'
  
  if board.b[newpos[0]][newpos[1]][0][0] == oppcolor:
    return board.b[newpos[0]][newpos[1]]
  else:
    return None

#Will check all possible surroundings to see if our piece can be attacked from anywhere. Will return true if it can be attacked and false if it it safe
def CanBeCaptured(pc,pos, board):
  owncolor = pc[0][0]
  oppcolor = '-'
  if owncolor == 'B':
    oppcolor = 'W'
  elif owncolor == 'W':
    oppcolor = 'B'
    
  #check up
  i = pos[0]
  j = pos[1]
  while i > 0:
    i -= 1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'R' or board.b[i][j][0][1] == 'Q' :
        #print("UP")
        return True
      break
  
  #check down
  i = pos[0]
  j = pos[1]
  while i < 7:
    i += 1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'R' or board.b[i][j][0][1] == 'Q' :
        #print("DOWN")
        return True
      break
  
  #check right
  i = pos[0]
  j = pos[1]
  while j < 7 :
    j += 1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'R' or board.b[i][j][0][1] == 'Q' :
        #print("RIGHT")
        return True
      break

  #check left
  i = pos[0]
  j = pos[1]
  while j > 0:
    j -= 1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'R' or board.b[i][j][0][1] == 'Q' :
        #print("left")
        return True
      break

  #check diagonals
  i = pos[0]
  j = pos[1]
  while i > 0 and j > 0:
    i -= 1
    j -=1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'B' or board.b[i][j][0][1] == 'Q' :
        #print("NW")
        return True
      break
  #check diagonals
  i = pos[0]
  j = pos[1]
  while i < 7 and j < 7:
    i += 1
    j +=1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'B' or board.b[i][j][0][1] == 'Q' :
        #print("SW")
        return True
      break
  #check diagonals
  i = pos[0]
  j = pos[1]
  while i > 0 and j < 7:
    i -= 1
    j +=1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'B' or board.b[i][j][0][1] == 'Q' :
        #print("SE")
        return True
      break
  #check diagonals
  i = pos[0]
  j = pos[1]
  while i < 7 and j > 0:
    i += 1
    j -=1
    if board.b[i][j][0][0] == owncolor:
      break
    if board.b[i][j][0][0] == oppcolor:
      if board.b[i][j][0][1] == 'B' or board.b[i][j][0][1] == 'Q' :
        #print("NE")
        return True
      break
  
  i = pos[0]
  j = pos[1]
  #check if a King can capture
  if i-1 > -1 and board.b[i-1][j][0][0] == oppcolor and board.b[i-1][j][0][1] == 'K':
    #print("King")
    return True
  if i+1 < 8 and board.b[i+1][j][0][0] == oppcolor and board.b[i+1][j][0][1] == 'K':
    #print("king")
    return True
  if j-1 > -1 and board.b[i][j-1][0][0] == oppcolor and board.b[i][j-1][0][1] == 'K':
    #print("king")
    return True
  if j+1 < 8 and board.b[i][j+1][0][0] == oppcolor and board.b[i][j+1][0][1] == 'K':
    #print("king")
    return True
  if i-1 > -1 and j-1 > -1 and board.b[i-1][j-1][0][0] == oppcolor and board.b[i-1][j-1][0][1] == 'K':
    #print("king")
    return True
  if i+1 < 8 and j+1 < 8 and board.b[i+1][j+1][0][0] == oppcolor and board.b[i+1][j+1][0][1] == 'K':
    #print("king")
    return True
  if i-1 > -1 and j+1 < 8 and board.b[i-1][j+1][0][0] == oppcolor and board.b[i-1][j+1][0][1] == 'K':
    #print("king")
    return True
  if i+1 < 8 and j-1 > -1 and board.b[i+1][j-1][0][0] == oppcolor and board.b[i+1][j-1][0][1] == 'K':
    #print("king")
    return True

  #check if a Pawn can capture
  if owncolor == 'W':
    if i+1 < 8 and j-1 > -1 and board.b[i+1][j-1][0][0] == oppcolor and board.b[i+1][j-1][0][1] == 'P':
     # print("pawn")
      return True
    if i+1 < 8 and j+1 < 8 and board.b[i+1][j+1][0][0] == oppcolor and board.b[i+1][j+1][0][1] == 'P':
    #  print("pawn")
      return True
  elif owncolor == 'B':
    if i-1 > -1 and j+1 < 8 and board.b[i-1][j+1][0][0] == oppcolor and board.b[i-1][j+1][0][1] == 'P':
      #print("pawn")
      return True
    if i-1 > -1 and j-1 < 8 and board.b[i-1][j-1][0][0] == oppcolor and board.b[i-1][j-1][0][1] == 'P':
      #print("pawn")
      return True
      
  #check if a knight can capture
  if i+2 < 8 and j-1 > -1 and board.b[i+2][j-1][0][0] == oppcolor and board.b[i+2][j-1][0][1] == 'N':
    #print("knight")
    return True
  if i+2 < 8 and j+1 < 8 and board.b[i+2][j+1][0][0] == oppcolor and board.b[i+2][j+1][0][1] == 'N':
    #print("knight")
    return True
  if i-2 > -1 and j-1 > -1 and board.b[i-2][j-1][0][0] == oppcolor and board.b[i-2][j-1][0][1] == 'N':
    #print("knight")
    return True
  if i-2 > -1 and j+1 < 8 and board.b[i-2][j+1][0][0] == oppcolor and board.b[i-2][j+1][0][1] == 'N':
    #print("knight")
    return True
  if j+2 < 8 and i-1 > -1 and board.b[i-1][j+2][0][0] == oppcolor and board.b[i-1][j+2][0][1] == 'N':
    #print("knight")
    return True
  if i+1 < 8 and j+2 < 8 and board.b[i+1][j+2][0][0] == oppcolor and board.b[i+1][j+2][0][1] == 'N':
    #print("knight")
    return True
  if j-2 > -1 and i-1 > -1 and board.b[i-1][j-2][0][0] == oppcolor and board.b[i-1][j-2][0][1] == 'N':
    #print("knight")
    return True
  if i+1 < 8 and j-2 > -1 and board.b[i+1][j-2][0][0] == oppcolor and board.b[i+1][j-2][0][1] == 'N':
    #print("knight")
    return True

  return False
```

```python
#If pawn moves to end of board then convert it to some other piece
def ConvertPawn(pos, board):
  i = pos[0]  #row
  j = pos[1]  #column
  if board.b[i][j] == BP:
    if i == 7:
      print("Enter which of the following you want to convert your pawn to:\n1. N - KNIGHT\n2. B - BISHOP\n3. Q - QUEEN\n4. R - ROOK")
      chose = input()
      if chose == 'N':
        board.b[i][j] = BN
      elif chose == 'B':
        board.b[i][j] = BB
      elif chose == 'Q':
        board.b[i][j] = BQ
      elif chose == 'R':
        board.b[i][j] = BR
  elif board.b[i][j] == WP:
    if i == 0:
      print("Enter which of the following you want to convert your pawn to:\n1. N - KNIGHT\n2. B - BISHOP\n3. Q - QUEEN\n4. R - ROOK")
      chose = input()
      if chose == 'N':
        board.b[i][j] = WN
      elif chose == 'B':
        board.b[i][j] = WB
      elif chose == 'Q':
        board.b[i][j] = WQ
      elif chose == 'R':
        board.b[i][j] = WR

#for computer randomly convert the pawn
def RandomlyConvertPawns(pos,board):
  choices = ['N','B','Q','R']
  i = pos[0]  #row
  j = pos[1]  #column
  if board.b[i][j] == BP:
    if i == 7:
      chose = random.choice(choices)
      if chose == 'N':
        board.b[i][j] = BN
      elif chose == 'B':
        board.b[i][j] = BB
      elif chose == 'Q':
        board.b[i][j] = BQ
      elif chose == 'R':
        board.b[i][j] = BR

  elif board.b[i][j] == WP:
    if i == 0:
      chose = random.choice(choices)
      if chose == 'N':
        board.b[i][j] = WN
      elif chose == 'B':
        board.b[i][j] = WB
      elif chose == 'Q':
        board.b[i][j] = WQ
      elif chose == 'R':
        board.b[i][j] = WR
```

```python
def EvaluateMove(oldpos, newpos, board):
  pc = board.b[oldpos[0]][oldpos[1]]              #get the piece we are trying to move
  pos_score = 0
  if pc[0] == 'WQ' or pc[0] == 'BQ':
    pos_score = QueenEval[newpos[0]][newpos[1]]
  elif pc[0] == 'WK':
    pos_score = WhiteKingEval[newpos[0]][newpos[1]]
  elif pc[0] == 'WR':
    pos_score = WhiteRookEval[newpos[0]][newpos[1]]
  elif pc[0] == 'WB':
    pos_score = WhiteBishopEval[newpos[0]][newpos[1]]
  elif pc[0] == 'WN' or pc[0] == 'BN':
    pos_score = KnightEval[newpos[0]][newpos[1]]
  elif pc[0] == 'WP':
    pos_score = WhitePawnEval[newpos[0]][newpos[1]]
  elif pc[0] == 'BK':
    pos_score = BlackKingEval[newpos[0]][newpos[1]]
  elif pc[0] == 'BR':
    pos_score = BlackRookEval[newpos[0]][newpos[1]]
  elif pc[0] == 'BB':
    pos_score = BlackBishopEval[newpos[0]][newpos[1]]
  elif pc[0] == 'BP':
    pos_score = BlackPawnEval[newpos[0]][newpos[1]]

  #weight of the opponent it can capture if it moves on new pos
  captureweight = CanCaptureOpponent(oldpos, newpos, board)
  if captureweight == None:
    captureweight = 0
  else:
    captureweight = captureweight[1]*3    #to add a little more weight to the capture
  #in case the piece can be captured from this position calculate its loss
  loss = 0
  if CanBeCaptured(pc,newpos, board):
    loss = pc[1]*2
  #if a piece is endangered in it current position
  protect = 0
  if CanBeCaptured(pc,oldpos,board):
    protect = pc[1]*2

  #print(captureweight)
  #print(loss)
  #print(protect)

  #When evaluating to decide how good a move is we see if the position is good on board, if we can capture an opponent, be captured ourselves and how necessary it is to move from current i.e whether we are unsafe currently
  if pc != WK and pc != BK:
    evaluation_score =  (pc[1]*pos_score)  + captureweight - loss + protect
  else: #change position weight of king because it will always be too much since it is 1000
    evaluation_score =  (10*pos_score)  + captureweight - loss + protect

  return evaluation_score
```

```python
#make sure a move is valid before making it
def valid_move(prev_pos, new_pos, board):
    moves = GetLegalMoves(prev_pos,board,8)
    if new_pos in moves:
      return True
    else:
      return False

#this is a non AI initial tester function for computer play
def computer_play(prev_pos, board):
  comp_moves = GetLegalMoves(prev_pos,board,8)
  while comp_moves == []:
    print("Please pick a position of a piece")
    prev_pos = input()
    prev_pos = convertPos(prev_pos)
    comp_moves = GetLegalMoves(prev_pos,board,8)
  comp_next = random.choice(comp_moves)
  if valid_move(prev_pos, comp_next, board):
    board.MovePiece(prev_pos, comp_next)
    RandomlyConvertPawns(comp_next,board)
    return True
  else:
    return False

#end game is King is Captured i.e Checkmate
def Game_Won(board):
  if 'WK' in captured:
      print("Check Mate")
      print("Computer Wins!!")
      return True
  elif 'BK' in captured:
    print("Check Mate")
    print("Player Wins!!")
    return True
  return False
```

```python
def GetAllCompPCPos(board):
  pcs = list()
  i = 0
  for i in range(8):
    j = 0
    for j in range(8):
      if board.b[i][j][0][0] == 'W':
        pcs.append([i,j])

  return pcs

def GetBestMovefromAPosition(curr_pos,board):
  allpossiblemoves = GetLegalMoves(curr_pos,board,100)
  BestMove = [[-1,-1],-999999]
  for eachmove in allpossiblemoves:
    getscore = EvaluateMove(curr_pos,eachmove,board)
    if getscore > BestMove[1]:
      BestMove[0] = eachmove
      BestMove[1] = getscore
  
  return BestMove
```

```python
def ComputerMove(board):
  All_Positions = GetAllCompPCPos(board)
  BestNextMoves = list()
  for each in All_Positions:
    tempmove = list()
    tempmove.append(each)
    temp = GetBestMovefromAPosition(each,board)
    for info in temp:
      tempmove.append(info)
    if tempmove[1] != [-1,-1]:
      BestNextMoves.append(tempmove)

  BestNextMoves.sort(key = lambda x: x[2])
  choosefrommoves = list()
  num = random.randint(1,2)
  if num == 1:
    if len(BestNextMoves)>2:
      i = len(BestNextMoves)-1
      while i > 4:
        choosefrommoves.append(BestNextMoves[i])
        i-=1
      chosemove = random.choice(choosefrommoves)
      return chosemove
    else:
      return BestNextMoves[len(BestNextMoves)-1]
  else:
    return BestNextMoves[len(BestNextMoves)-1]
```
